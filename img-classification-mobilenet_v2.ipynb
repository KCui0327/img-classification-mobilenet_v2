```python
import tensorflow as tf
import numpy as np
img_file = "C:/Users/Kencu/OneDrive/Desktop/img-classification/test.jpg"
```

```python
from tensorflow.keras.preprocessing import image as img_class
img = img_class.load_img(img_file, target_size = (224, 224))
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)S
```

```python
mb_net = tf.keras.applications.mobilenet_v2.MobileNetV2()
```

```python
resize_test_img = img_class.img_to_array(img)
final_img = np.expand_dims(resize_test_img, axis=0)
final_img = tf.keras.applications.mobilenet_v2.preprocess_input(final_img)
```

```python
from tensorflow.keras.applications import imagenet_utils as img_utils
predict = mb_net.predict(final_img)
result = img_utils.decode_predictions(predict)
```
